# Sorting

```python
import numpy as np
import pandas as pd
```

```python
df=pd.DataFrame(np.random.randn(10,2),index=[1,4,6,2,3,5,9,8,0,7],
                columns=['col2','col1'])
df
```

### Two methods
1. By Lable [sorting indexs and columns]
2. By Actual values

#### Sorting using Lable

```python
df
```

```python
df = df.sort_index() #Index gets sorted 
df
```

```python
df = df.sort_index(axis =1) # Columns gets sorted
df
```

```python
df = df.sort_index(ascending=False)
```

```python
df
```

#### Sorting using Values

```python
df
```

```python
df = df.sort_values('col1')
df
```

```python
df = df.sort_values(by='col2')
df
```

```python
df = df.sort_values(by = 'col1', ascending=False)
df
```

```python
df = df.sort_values(by = 'col1')
df
```
